```python
def dpar(M):
    dic = {}

    for num in M:
        if num not in dic:
            dic[num] = 1

        else:
            dic[num] += 1

    return dic
```

```python
print(dpar([12, 19, 19, 18, 18, 16, 18, 13, 19, 18, 12, 18, 11, 20, 14, 14, 20, 20, 20, 16, 13, 15, 19, 14, 12]))
```

```python
def TempMaxMin(lst):
    dic = {}

    for ciudad in lst:
        nomCiudad = ciudad[0]
        tempCiudad = ciudad[1:]
        
        dic[nomCiudad] = [max(tempCiudad), min(tempCiudad)]

    return dic
```

```python
print(TempMaxMin([['Londres', 3.4, 6.3, 10.5, 6.8], \
        ['Oslo', -3.8, -5.0, 5.1, 4.2], ['Berlin', 7.5, 4.1, 12.3, 13.0], \
        ['Málaga', 14.7, 12.3, 19.5, 18.4]]))
```

```python
def PersMayTemp(dPersC, dCiudT):
    residentes = []

    for persona in dPersC:
        ciudad = dPersC[persona]
        temperatura = dCiudT[ciudad]
        
        if temperatura < 0:
            residentes.append(persona)

    residentes.sort()

    return residentes
```

```python
def lst_hiper(dic, edad):
    hiperT = []

    for persona in dic:
        datos = dic[persona]
        edadN = datos[0]
        sis = datos[1]
        dia = datos[2]

        if (sis >= 140 or dia >= 90) and edadN < edad:
            hiperT.append(persona)

    hiperT.sort()

    return hiperT
```

```python
def nivelKsang(dK, lst):
    resultados = {}

    for paciente in dK:
        nivelK = dK[paciente]

        if nivelK < lst[0]:
            resultados[paciente] = 'hipokalemia crítica'

        if lst[0] <= nivelK < lst[1]:
            resultados[paciente] = 'hipokalemia leve'

        if lst[1] <= nivelK <= lst[2]:
            resultados[paciente] = 'normal'

        if lst[2] < nivelK <= lst[3]:
            resultados[paciente] = 'hiperkalemia moderada'

        if nivelK > lst[3]:
            resultados[paciente] = 'hiperkalemia severa'
    
    return resultados
```

```python
dK1 = {'Luis': 2.2, 'Carlos': 7.0, 'Laia': 4.0, 'Mikel': 5.5, \
               'Jordi': 5.2, 'Anna': 3.6, 'Joe': 7.2}
ls1 = [2.0, 3.5, 5.2, 7.0]
print(nivelKsang(dK1, ls1))
```

```python
import pandas as pd

lst_ciudad = [['Londres', 3.4, 6.3, 10.5, 6.8], ['Oslo', -3.8, -5.0, 5.1, 4.2], ['Berlin', 7.5, 4.1, 12.3, 13.0], ['Málaga', 14.7, 12.3, 19.5, 18.4]]
df = pd.DataFrame(lst_ciudad, columns = ['Ciudad', 'Enero', 'Febrero', 'Marzo', 'Abril'])
df.name = 'Temperatura ciudades'

print(df)
```

```python
import pandas as pd

def df_hiper(dhiper):
    # Defino variables como listas vacías
    nombres = []
    edades = []
    sist = []
    diast = []
    diagnos = []

    # Relleno las listas vacías con los nombres y datos del diccionario de entrada
    for persona in dhiper:
        datos = dhiper[persona]
        nombres.append(persona)
        edades.append(datos[0])
        sist.append(datos[1])
        diast.append(datos[2])

        # Creo la columna de diagnóstico con las condiciones del enunciado. 
        # datos[1] es el valor de la presión sistólica que leo en ese momento, y datos[2] el de la diastólica
        if datos[1] < 90 or datos[2] < 60:
            diagnos.append("baja")

        elif datos[1] >= 140 or datos[2] >= 90:
            diagnos.append("alta")

        else:
            diagnos.append("normal")

    # Creo un nuevo diccionario con las listas creadas anteriormente, con claves "Nombre", "Edad", "Sistólica", "Diastólica", "Diagnóstico"
    dhiper_df = {"Nombre": nombres, "Edad": edades, "Sistólica": sist, "Diastólica": diast, "Diagnóstico": diagnos}

    return pd.DataFrame(dhiper_df)
```

```python
dpers = {'Maria': [40, 135, 90],'Nuria': [63, 141, 92], \
                 'Jose': [47, 110, 59], 'Luis': [49, 146, 94], \
                 'Oriol': [52, 130, 89], 'Carlos': [65, 125, 89], \
                 'Pepe': [70, 130, 92]}
print(df_hiper(dpers))
```
